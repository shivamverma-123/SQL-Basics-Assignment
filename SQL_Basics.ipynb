Q1. Create a New Database and  Table for Employees

Task: Create a new database named Company_db and Create a table named employees with the following

Ans: To create a new database and table for emplyoyees here is code for this.

-- Step 1: Create Database
CREATE DATABASE company_db;

-- Step 2: Use the Database
USE company_db;

-- Step 3: Create Employees Table
CREATE TABLE employees (

    employee_id INT PRIMARY KEY,

    first_name VARCHAR(50),
    last_name VARCHAR(50),
    department VARCHAR(50),
    salary INT,
    hire_date DATE
);

Explanation (short):

- company_db → new database create ki

- employees → table ka naam

- employee_id → PRIMARY KEY

- VARCHAR(50) → text ke liye

- DATE → date store karne ke liye

Q3.Display All Employee Records Sorted by Salary (Lowest to Highest)

Hint: Use the ORDER BY clause on the salary column.

Ans:Using order by clause to sorted by salary column in (lowest to higest )

SELECT *
FROM employees
ORDER BY salary ASC;

Explanation (short):

- ORDER BY salary → salary ke basis par sort

- ASC → Lowest to Highest (ascending order)

- ASC optional hota hai (by default ascending hi hota hai)

Q9.What is the Difference Between a Primary Key and a Foreign Key?

Ans:The difference between a primary key and foregin key are as follow as.


1.Primary Key table ke har record ko uniquely identify karti hai, jabki.

 Foreign Key ek table ke column ko dusre table ki Primary Key se link karti hai.

2.Primary Key me duplicate values allowed nahi hoti, jabki

Foreign Key me duplicate values ho sakti hain.

3.Primary Key ka value NULL nahi ho sakta, jabki

Foreign Key me NULL allowed hota hai.

4.Primary Key same table ke andar hoti hai, jabki

Foreign Key do tables ke beech relationship banati hai.

5.Primary Key ka main purpose data uniqueness maintain karna hota hai, jabki

Foreign Key ka purpose referential integrity maintain karna hota hai.

6.Ek table me sirf ek Primary Key hoti hai, jabki multiple Foreign Keys ho sakti hain.

7.Primary Key ko reference karke hi Foreign Key banayi ja sakti hai, lekin Foreign Key bina Primary Key ka kaam nahi karti.    

Q10. What Are Constraints in SQL and Why Are They Used?

Ans: The constraints in SQL are

Constraints define conditions that data must satisfy before it can be inserted, updated, or deleted in a table.
They help ensure that only valid data is stored in the database.

they are used are as follow as

1.Maintain data accuracy:

→ Prevent invalid or incorrect values from being stored.

2.Ensure data integrity:

→ Keep relationships between tables consistent.

3.Avoid duplicate records:

→ Ensure unique values where required.

4.Enforce business rules:

→ Example: salary must be greater than zero.

5.Protect database reliability:

→ Reduces errors during data entry and updates.
